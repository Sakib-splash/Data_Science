Task 1: Dataset Preparation
You can download the dataset from the provided Kaggle link, or generate synthetic data.

```python
import numpy as np
import pandas as pd

# Generate synthetic data (if needed)
np.random.seed(42)
X = np.random.rand(30) * 100  # Independent variable
Y = 3.5 * X + np.random.randn(30) * 20  # Dependent variable with some noise

# Convert to DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Display the first few rows
print(data.head())
```

Task 2: Data Visualization
Visualize the relationship between X and Y using a scatter plot.

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.6)
plt.title('Scatter Plot of X vs. Y')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.grid(True)
plt.show()
```

Task 3: Linear Regression Implementation
Use scikit-learn to perform linear regression and calculate the slope and intercept.

```python
from sklearn.linear_model import LinearRegression

# Reshape X for the model
X_reshaped = data['X'].values.reshape(-1, 1)

# Initialize and fit the model
model = LinearRegression()
model.fit(X_reshaped, data['Y'])

# Extract slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
```

Task 4: Coefficient of Determination (R²)
Calculate and explain R2 , the proportion of variance explained by the model.

```python
# Calculate R²
r_squared = model.score(X_reshaped, data['Y'])

print(f"Coefficient of Determination (R²): {r_squared}")
```

Explanation:
𝑅2 measures how well the model explains the variability of the dependent variable. An R2 close to 1 indicates a strong relationship.

Task 5: Predictions and Visualization
Predict
𝑌 and overlay the regression line on the scatter plot.

```python
# Predictions
Y_pred = model.predict(X_reshaped)

# Plot the original data and regression line
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.6, label='Original Data')
plt.plot(data['X'], Y_pred, color='red', label='Regression Line')
plt.title('Regression Line Overlay')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

Task 6: Residual Analysis
Analyze residuals to check for patterns.

```python
# Calculate residuals
residuals = data['Y'] - Y_pred

# Plot residuals
plt.scatter(data['X'], residuals, color='purple', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', label='Zero Error Line')
plt.title('Residual Plot')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Residuals')
plt.legend()
plt.grid(True)
plt.show()

# Interpretation
print("Residual analysis: Look for random patterns. Patterns indicate model assumption violations.")
```

Task 7: Model Evaluation
Compute evaluation metrics: MAE, MSE, and RMSE.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Metrics calculation
mae = mean_absolute_error(data['Y'], Y_pred)
mse = mean_squared_error(data['Y'], Y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

Explanation:

MAE measures the average magnitude of errors.
MSE gives more weight to larger errors due to squaring.
RMSE is the standard deviation of prediction errors.
